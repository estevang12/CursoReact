## 1. Planificación de Proyecto, Diseño y Arquitectura

Aquí están los prompts extraídos formateados en cajas de código:

**Para configurar un nuevo proyecto:**
```
Estoy comenzando un nuevo [tipo de proyecto] usando [lenguaje de programación/marco de trabajo].
¿Puedes sugerir una estructura básica de archivos y dependencias esenciales que debería considerar?
```

**Para entender las capacidades de la IA:**
```
¿Cuáles son las principales fortalezas y limitaciones que debo tener en cuenta
al usar tu ayuda para tareas de desarrollo de software?
```

**Para crear un plan de codificación:**
```
Quiero construir [breve descripción del proyecto]. ¿Puedes ayudarme a desglosarlo
en tareas manejables y sugerir un orden de implementación?
```

**Creando una base de conocimiento para tu asistente de IA:**
```
Estoy comenzando un nuevo proyecto y me gustaría crear una base de conocimiento para que puedas referenciar durante nuestro proceso de desarrollo.
El proyecto es [breve descripción].
Usaremos [pila tecnológica].
Las características clave que necesitamos implementar son [listar características].
Algunas restricciones importantes a tener en cuenta son [listar restricciones].
En términos de estilo de codificación, preferimos [mencionar preferencias].
¿Puedes resumir esta información y sugerir detalles adicionales que deberíamos incluir en nuestra base de conocimiento?
```

**Desglosando proyectos en componentes manejables:**
```
Basándote en la visión general del proyecto que acabamos de crear, ¿puedes ayudarme a desglosar este proyecto en componentes o módulos manejables? Para cada componente, por favor sugiere:
1. Un nombre para el componente
2. Su funcionalidad principal
3. Desafíos potenciales en su implementación
4. Cómo podría interactuar con otros componentes
Por favor, también recomienda un orden lógico para desarrollar estos componentes.
```

**Usando IA para el plan de proyecto:**
```
Utilizando el desglose de componentes que hemos creado, ¿puedes ayudarme a desarrollar
una hoja de ruta del proyecto? Por favor incluye:
1. Un orden sugerido para desarrollar los componentes
2. Marcos de tiempo estimados para cada componente (asumiendo que trabajo en esto a tiempo parcial)
3. Posibles hitos o puntos de control
4. Cualquier dependencia entre componentes que pueda afectar el orden de desarrollo
5. Sugerencias para cualquier etapa de prueba de concepto o prototipo que pueda ser beneficiosa
```

**Para evaluación de riesgos:**
```
Basándonos en nuestro plan de proyecto, ¿puedes identificar posibles riesgos o desafíos
que podamos enfrentar durante el desarrollo? Para cada riesgo, sugiere posibles estrategias de mitigación.
```

**Para selección de tecnología:**
```
Estamos considerando usar [Tecnología A] o [Tecnología B]
para [funcionalidad específica]. ¿Puedes comparar estas opciones
en el contexto de nuestro proyecto, considerando factores como rendimiento,
facilidad de implementación y escalabilidad futura?
```

**Para estimar el alcance del proyecto:**
```
Dadas nuestras necesidades del proyecto y el desglose de componentes, ¿puedes ayudarme
a estimar el alcance general de este proyecto? Por favor, proporciona una estimación aproximada
del tiempo total de desarrollo y sugiere cualquier área donde podríamos necesitar
ajustar nuestras expectativas o buscar recursos adicionales.
```

**Para diseño y arquitectura:**
```
Pensemos paso a paso sobre la arquitectura para [componente o
sistema específico]. Por favor considera:
1. Las funcionalidades principales que este componente necesita soportar
2. Posibles estructuras de datos o modelos
3. Clases o módulos clave y sus responsabilidades
4. Cómo este componente interactuará con otras partes del sistema
5. Patrones de diseño que podrían ser aplicables
6. Consideraciones para escalabilidad y mantenibilidad

Para cada paso, proporciona una breve explicación de tu razonamiento.
```

**Para iterar sobre las sugerencias de IA:**
```
Gracias por ese diseño inicial. Tengo algunas preguntas de seguimiento:
1. ¿Cuáles son los posibles inconvenientes o limitaciones de este enfoque?
2. ¿Puedes sugerir un diseño alternativo que priorice [preocupación específica,
por ejemplo, rendimiento, flexibilidad]?
3. ¿Cómo cambiaría este diseño si necesitáramos [requisito futuro
potencial]?
```

**Para documentar decisiones de arquitectura:**
```
Basándonos en nuestra discusión de la arquitectura del [componente/sistema], ¿puedes ayudarme
a crear un Registro de Decisiones de Arquitectura (ADR)? Por favor incluye:
1. El contexto y problema que estamos abordando
2. Las opciones que consideramos
3. La decisión que tomamos
4. Las consecuencias (tanto positivas como negativas) de esta decisión
5. Cualquier decisión o compromiso relacionado
```

**Para explorar patrones de diseño:**
```
Dada nuestra necesidad de [funcionalidad específica], ¿qué patrones de diseño
podrían ser aplicables? Para cada patrón sugerido, por favor explica cómo podría
implementarse en nuestro sistema y qué beneficios proporcionaría.
```

**Para diseño de esquema de base de datos:**
```
Necesitamos diseñar un esquema de base de datos para [parte específica del sistema]. Basado
en nuestros requisitos, ¿puedes sugerir un diseño de esquema inicial? Por favor incluye
tablas, campos clave y relaciones. Además, considera posibles estrategias de indexación
para mejorar el rendimiento.
```

**Para diseño de API:**
```
Estamos planeando crear una API RESTful para [funcionalidad específica]. ¿Puedes
ayudar a diseñar los endpoints que necesitaremos? Para cada endpoint, sugiere el método HTTP,
estructura de URL, formatos de solicitud/respuesta y cualquier requisito de autenticación.
```

**Para consideraciones de escalabilidad:**
```
Al diseñar nuestro sistema, necesitamos asegurarnos de que pueda escalar para manejar [carga
esperada]. ¿Puedes revisar nuestra arquitectura actual y sugerir modificaciones o
componentes adicionales que podríamos necesitar para asegurar la escalabilidad? Por favor considera
estrategias de escalado tanto vertical como horizontal.
```

## 2. Generación de Código, Diseño de Bases de Datos y Documentación

### Dominando la Generación de Código Asistida por IA

**Creando Prompts Efectivos para la Generación de Código**
```
Necesito implementar [funcionalidad específica] en [lenguaje de programación].
Requisitos clave:
1. [Requisito 1]
2. [Requisito 2]
3. [Requisito 3]
Por favor considera:
- Manejo de errores
- Casos límite
- Optimización del rendimiento
- Mejores prácticas para [lenguaje/marco de trabajo]
Por favor, no elimines innecesariamente ningún comentario o código.
Genera el código con comentarios claros que expliquen la lógica.
```

**Revisando y Entendiendo el Código Generado por IA**
```
¿Puedes explicar en detalle la siguiente parte del código:
[pegar sección de código]
Específicamente:
1. ¿Cuál es el propósito de esta sección?
2. ¿Cómo funciona paso a paso?
3. ¿Hay posibles problemas o limitaciones con este enfoque?
```

**Aprovechando la IA para Revisiones e Mejoras de Código**
```
Por favor revisa el siguiente código:
[pegar tu código]
Considera:
1. Calidad del código y adhesión a las mejores prácticas
2. Posibles bugs o casos límite
3. Optimización del rendimiento
4. Legibilidad y mantenibilidad
5. Cualquier preocupación de seguridad
Sugiere mejoras y explica tu razonamiento para cada sugerencia.
```

**Ideas de Prompts para Varias Tareas de Codificación**

*Para implementar un algoritmo específico:*
```
Implementa un [nombre del algoritmo] en [lenguaje de programación].
Por favor incluye:
1. La función principal con tipos de parámetros y retorno claros
2. Funciones auxiliares si es necesario
3. Análisis de complejidad de tiempo y espacio
4. Ejemplo de uso
```

*Para crear una clase o módulo:*
```
Crea una [clase/módulo] para [funcionalidad específica] en [lenguaje de programación].
Incluye:
1. Constructor/inicialización
2. Métodos principales con docstrings claros
3. Cualquier método auxiliar privado necesario
4. Encapsulación adecuada y adhesión a principios de POO
```

*Para optimizar código existente:*
```
Aquí hay un código que necesita optimización:
[pegar código]
Por favor sugiere optimizaciones para mejorar su rendimiento. Para cada sugerencia, explica la mejora esperada y cualquier compensación.
```

### Diseño de Bases de Datos y Optimización de Consultas con IA

**Aprovechando la IA para Esquemas de Bases de Datos**
```
Estoy diseñando una base de datos para [describe tu aplicación]. Las entidades principales son:
[Lista de entidades principales]

Requisitos clave:
1. [Requisito 1, por ejemplo, "Debe soportar recuperación rápida de publicaciones de usuarios"]
2. [Requisito 2, por ejemplo, "Necesita rastrear relaciones de usuarios (seguidores/siguiendo)"]
3. [Requisito 3, por ejemplo, "Debe manejar grandes volúmenes de datos de series temporales para análisis"]

Por favor sugiere un esquema de base de datos que incluya:
1. Tablas y sus columnas (con tipos de datos apropiados)
2. Relaciones de claves primarias y foráneas
3. Cualquier tabla de unión necesaria para relaciones muchos a muchos
4. Índices sugeridos para el rendimiento
5. Consideraciones para escalabilidad

Además, por favor explica el razonamiento detrás de tus decisiones de diseño.
```

**Refinando el Esquema**
```
Dado este esquema inicial:
[Pegar tu esquema aquí]

Por favor analízalo considerando lo siguiente:
1. Normalización: ¿El esquema está correctamente normalizado? Si no, sugiere mejoras.
2. Desnormalización: ¿Hay casos donde la desnormalización podría mejorar el rendimiento?
3. Estrategia de indexación: Sugiere índices adicionales que podrían mejorar el rendimiento de las consultas.
4. Escalabilidad: ¿Cómo manejará este esquema el crecimiento? ¿Algún posible cuello de botella?
5. Integridad de datos: ¿Hay restricciones o triggers que deberíamos considerar para asegurar la consistencia de datos?
Para cada sugerencia, por favor explica los pros y contras.
```

**Creando Consultas Optimizadas**
```
Necesito optimizar la siguiente consulta SQL:
[Pegar tu consulta aquí]

La consulta está tardando demasiado en ejecutarse en conjuntos de datos grandes. Por favor:
1. Analiza la consulta e identifica posibles problemas de rendimiento.
2. Sugiere optimizaciones, que pueden incluir:
   - Reescribir la consulta
   - Añadir o modificar índices
   - Sugerencias para cambios en el esquema si es necesario
3. Explica el razonamiento detrás de cada optimización.
4. Si es posible, proporciona una estimación de la mejora de rendimiento que podríamos esperar.
Contexto adicional:
- Sistema de base de datos: [por ejemplo, PostgreSQL, MySQL]
- Tamaños aproximados de tablas: [por ejemplo, la tabla Usuarios tiene 1 millón de filas]
- Cualquier restricción de hardware relevante
```

**Ideas de Prompts para Tareas Relacionadas con Bases de Datos**

*Para optimización de índices:*
```
Dada la siguiente estructura de tabla y consultas comunes:
[Pegar estructura de tabla y consultas de ejemplo]

Por favor sugiere una estrategia de indexación óptima. Considera:
1. ¿Qué columnas deberían ser indexadas?
2. ¿Deberíamos usar índices de columna única o multicolumna?
3. ¿Hay casos donde un índice cubriente sería beneficioso?
4. ¿Cómo podrían estos índices afectar el rendimiento de escritura?
```

*Para planificación de migración de datos:*
```
Necesito migrar datos de un esquema antiguo a uno nuevo:
Esquema antiguo:
[Pegar esquema antiguo]

Esquema nuevo:
[Pegar esquema nuevo]

Por favor ayúdame a crear un plan de migración:
1. Identifica los pasos necesarios para transformar los datos
2. Sugiere tablas o vistas intermedias que podrían ser útiles
3. Considera la integridad de datos y cómo manejar posibles conflictos
4. Propón una estrategia para verificar el éxito de la migración
```

*Para solución de problemas de rendimiento de consultas:*
```
La siguiente consulta está teniendo un rendimiento deficiente:
[Pegar consulta problemática]

Plan de ejecución:
[Pegar plan de ejecución si está disponible]

Por favor analiza esta consulta y sugiere mejoras:
1. Identifica cualquier parte subóptima de la consulta o plan de ejecución
2. Sugiere formas alternativas de escribir la consulta
3. ¿Hay índices faltantes que podrían ayudar?
4. ¿Ayudaría desnormalizar alguna parte del esquema para mejorar el rendimiento de esta consulta?
```

### Documentación Hecha Fácil

**Generando Documentación con IA**
```
Necesito crear documentación para [nombre del proyecto/componente]. Por favor genera:

1. Una visión general del [proyecto/componente]
2. Instrucciones de instalación
3. Opciones de configuración
4. Referencia de API (si aplica)
5. Ejemplos de uso
6. Guía de resolución de problemas
7. Sección de preguntas frecuentes

Para cada sección, considera:
- El público objetivo (por ejemplo, desarrolladores, usuarios finales)
- Cualquier requisito previo o dependencia
- Obstáculos comunes o conceptos erróneos
- Mejores prácticas

Por favor usa un lenguaje claro y conciso e incluye fragmentos de código relevantes donde sea apropiado.
```

**Refinando Documentos Generados por IA para Precisión y Claridad**
```
Por favor revisa y mejora la siguiente sección de documentación:
[Pegar sección aquí]

Considera:
1. Claridad de la explicación
2. Compleción de la información
3. Nivel de detalle apropiado para el público objetivo
4. Consistencia con las mejores prácticas en escritura técnica
Sugiere mejoras y explica tu razonamiento.
```

**Ideas de Prompts para Diferentes Tipos de Documentación**

*Para documentación de API:*
```
Genera documentación de API para el siguiente endpoint:
[Pegar detalles del endpoint]
Incluye:
1. URL del endpoint y método
2. Parámetros de solicitud y sus tipos
3. Formato del cuerpo de la solicitud (si aplica)
4. Formato de respuesta y posibles códigos de estado
5. Ejemplo de solicitud y respuesta
6. Cualquier requisito de autenticación
7. Información de limitación de velocidad (si aplica)
```

*Para un archivo README:*
```
Crea un archivo README.md para mi repositorio de GitHub. El proyecto es [breve descripción]. Incluye:
1. Título y descripción del proyecto
2. Instrucciones de instalación
3. Ejemplos de uso
4. Guías de contribución
5. Información de licencia
6. Insignias (por ejemplo, estado de construcción, versión, etc.)
Usa formato Markdown adecuado y considera añadir una tabla de contenido para una navegación más fácil.
```

*Para guías de usuario:*
```
Genera una guía de usuario para [nombre del producto/función]. El público objetivo es [describir audiencia]. Incluye:
1. Introducción y propósito del producto/función
2. Guía de inicio rápido
3. Funciones principales y cómo usarlas
4. Consejos de uso avanzado
5. Resolución de problemas de problemas comunes
Usa lenguaje simple y considera incluir instrucciones paso a paso con marcadores de posición para capturas de pantalla hipotéticas.
```

*Para comentarios de código y docstrings:*
```
Genera comentarios y docstrings apropiados para el siguiente código:
[Pegar código aquí]
Sigue las convenciones de docstrings específicas de [lenguaje]. Incluye:
1. Breve descripción de la función/clase
2. Parámetros y sus tipos
3. Valor de retorno y tipo
4. Cualquier excepción que pueda ser lanzada
5. Ejemplos de uso si el uso de la función/clase no es inmediatamente obvio
```

**Aprovechando la IA para Documentación Viva**
```
He realizado los siguientes cambios en mi código:
[Resumir cambios]
Por favor actualiza las secciones relevantes de la documentación para reflejar estos cambios.
Destaca cualquier cambio disruptivo o nuevas características que los usuarios deban conocer.
```
```
Por favor revisa la siguiente documentación:
[Pegar documentos actuales]
Considerando las últimas mejores prácticas y puntos de dolor comunes de los usuarios en proyectos similares:
1. Sugiere cualquier sección que deba ser añadida o ampliada
2. Identifica cualquier parte que pueda estar desactualizada o ya no sea relevante
3. Recomienda mejoras para la claridad y facilidad de uso
```

## 3. Pruebas, Optimización y Control de Versiones

### Generando Pruebas Unitarias con IA
1. **Prompt para Pruebas Unitarias:**
   ```
   Necesito pruebas unitarias para la siguiente función:
   [Pegar tu función aquí]

   Por favor genera un conjunto integral de pruebas unitarias que cubran:
   1. Escenarios de ruta feliz
   2. Casos límite
   3. Condiciones de error
   4. Análisis de valores límite

   Para cada caso de prueba, por favor:
   1. Proporciona una breve descripción de lo que está comprobando la prueba
   2. Escribe el código real de la prueba usando [marco de pruebas preferido, por ejemplo, pytest]
   3. Explica cualquier objeto simulado o fixture que pueda ser necesario

   Además, sugiere cualquier prueba adicional que pueda ser relevante basándose en trampas comunes o mejores prácticas para este tipo de función.
   ```

### Técnicas de Depuración Asistidas por IA
2. **Prompt para Asistencia en Depuración:**
   ```
   Estoy encontrando el siguiente error:
   [Describir el error, incluyendo cualquier mensaje de error y los pasos para reproducirlo]

   Aquí está el código relevante:
   [Pegar el código relacionado con el error]

   Por favor ayúdame a depurar este problema:
   1. Analiza el código y sugiere posibles causas del error
   2. Proporciona estrategias de depuración paso a paso que pueda seguir
   3. Sugiere cualquier herramienta o técnica que pueda ser útil para diagnosticar el problema
   4. Si es posible, propone posibles soluciones y explica su razonamiento

   Además, ¿hay mejores prácticas o trampas comunes relacionadas con este tipo de problema que debería conocer para referencia futura?
   ```

### Mejora Continua a través de Revisiones de Código Asistidas por IA
3. **Prompt para Revisión de Código:**
   ```
   Por favor revisa el siguiente código en busca de calidad y posibles problemas:
   [Pegar tu código aquí]

   En tu revisión, por favor considera:
   1. Estilo de código y adhesión a las mejores prácticas
   2. Posibles bugs o casos límite no manejados
   3. Optimizaciones de rendimiento
   4. Vulnerabilidades de seguridad
   5. Legibilidad y mantenibilidad

   Para cada problema encontrado, por favor:
   1. Explica el problema
   2. Sugiere una solución
   3. Proporciona una breve justificación del cambio sugerido

   Además, ¿hay alguna mejora general o sugerencia de refactorización que harías para este código?
   ```

### Ideas de Prompts para Pruebas y Depuración
4. **Generando Pruebas de Integración:**
   ```
   Necesito crear pruebas de integración para los siguientes componentes:
   [Listar componentes y sus interacciones]

   Por favor sugiere un conjunto de pruebas de integración que:
   1. Cubran los principales escenarios de interacción entre estos componentes
   2. Prueben el manejo adecuado de errores y casos límite
   3. Incluyan cualquier procedimiento necesario de configuración y desmontaje

   Proporciona los escenarios de prueba en un formato claro, paso a paso, e incluye cualquier objeto simulado o datos de prueba necesarios.
   ```

5. **Pruebas de Rendimiento:**
   ```
   Necesito crear un plan de pruebas de rendimiento para mi aplicación. Las áreas clave de preocupación son:
   [Listar funcionalidades o componentes principales a probar]

   Por favor ayúdame a crear un plan de pruebas de rendimiento que incluya:
   1. Indicadores clave de rendimiento a medir
   2. Escenarios de prueba para simular varias condiciones de carga
   3. Sugerencias de herramientas o marcos a utilizar
   4. Estrategias para identificar cuellos de botella de rendimiento
   5. Mejores prácticas para interpretar y actuar sobre los resultados
   ```

6. **Pruebas de Seguridad:**
   ```
   Por favor revisa el siguiente código en busca de posibles vulnerabilidades de seguridad:
   [Pegar tu código]

   Considera problemas de seguridad comunes como:
   1. Fallos de inyección
   2. Autenticación rota
   3. Exposición de datos sensibles
   4. Entidades externas de XML (XXE)
   5. Control de acceso roto

   7. Cross-site scripting (XSS)

   Para cada vulnerabilidad encontrada, explica el riesgo y sugiere prácticas de codificación segura para mitigarlo.
   ```

7. **Generación de Datos de Prueba:**
   ```
   Necesito generar datos de prueba para el siguiente esquema de base de datos:
   [Pegar tu esquema aquí]

   Por favor ayúdame a crear un plan de generación de datos de prueba:
   1. Sugiere rangos o tipos de valores apropiados para cada campo
   2. Proporciona SQL o script para generar un conjunto diverso de datos de prueba, incluyendo:
      - Casos normales
      - Casos límite
      - Datos inválidos para probar el manejo de errores
   3. Asegura que se mantiene la integridad referencial para tablas relacionadas
   4. Incluye cualquier escenario específico o patrón de datos crucial para pruebas exhaustivas
   ```

### Usando IA para Mejorar la Seguridad del Código
8. **Auditoría de Seguridad:**
   ```
   Por favor realiza una auditoría de seguridad en el siguiente código:
   [Pegar tu código aquí]

   En tu auditoría, por favor:
   1. Identifica cualquier posible vulnerabilidad de seguridad, incluyendo pero no limitado a:
      - Fallos de inyección
      - Autenticación rota
      - Exposición de datos sensibles
      - Entidades Externas de XML (XXE)
      - Control de acceso roto
      - Configuraciones de seguridad incorrectas
      - Cross-Site Scripting (XSS)
      - Deserialización insegura
      - Uso de componentes con vulnerabilidades conocidas
      - Registro y monitoreo insuficientes
   2. Para cada vulnerabilidad encontrada:
      - Explica el impacto potencial
      - Sugiere una solución o estrategia de mitigación
      - Proporciona un fragmento de código demostrando la solución, si es aplicable
   3. Sugiere cualquier mejora general de seguridad o mejores prácticas que podrían aplicarse a este código.
   4. Recomienda cualquier biblioteca o herramienta relacionada con la seguridad que podría ayudar a mejorar la postura de seguridad general de la aplicación.
   ```

### Optimizando el Rendimiento con Sugerencias de IA
9. **Optimización de Rendimiento:**
   ```
   Por favor analiza el siguiente código en busca de oportunidades de optimización de rendimiento:
   [Pegar tu código aquí]

   En tu análisis, por favor:
   1. Identifica cualquier cuello de botella de rendimiento u operaciones ineficientes
   2. Sugiere optimizaciones, considerando:
      - Mejoras en la complejidad de tiempo
      - Optimización del uso de memoria
      - Reducción de operaciones o llamadas a funciones innecesarias
      - Potencial para paralelización u operaciones asíncronas
      - Estrategias de almacenamiento en caché
   3. Para cada sugerencia:
      - Explica el impacto esperado en el rendimiento
      - Proporciona un fragmento de código demostrando la optimización
      - Discute cualquier compensación potencial (por ejemplo, legibilidad, mantenibilidad)
   4. Recomienda cualquier mejor práctica específica del lenguaje o bibliotecas que podrían ser beneficiosas
   5. Sugiere cualquier herramienta o técnica de perfilado que podría ayudar a analizar más el rendimiento en un escenario del mundo real
   ```

### Control de Versiones y Colaboración con IA
10. **Mensajes de Commit:**
    ```
    He realizado los siguientes cambios en mi código:
    [Pegar tu diff de git o describir los cambios]

    Por favor ayúdame a crear un mensaje de commit que:
    1. Resuma los cambios de manera concisa (50 caracteres o menos para la línea de asunto)
    2. Proporcione más detalles en el cuerpo (ajustar a 72 caracteres)
    3. Siga las mejores prácticas para mensajes de commit de git
    4. Incluya cualquier número de problema o referencia relevante

    El mensaje de commit debe ser lo suficientemente informativo para que los miembros del equipo puedan entender los cambios sin tener que mirar el código.
    ```
